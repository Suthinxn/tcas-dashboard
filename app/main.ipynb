## Import library

```python
import pandas
```

## Load Data

```python
data = pandas.read_csv('data/courses.csv')
data.head(10)
```

```python
data.columns
```

## 

```python
filtered_ai_data = data[
    (data['program_name_th'].str.contains('วิศวกรรมปัญญาประดิษฐ์', case=False, na=False)) |
    (data['field_name_th'].str.contains('วิศวกรรมปัญญาประดิษฐ์', case=False, na=False))

]
filtered_ai_data.head(10)
```

```python
filtered_ai_data.to_excel('data/filtered_ai_courses.xlsx', index=False)
```

```python
filtered_com_data = data[
    (data['field_name_th'].str.contains('วิศวกรรมคอมพิวเตอร์', case=False, na=False))
]
filtered_com_data.head(10)
```

```python
filtered_com_data.to_excel('data/filtered_com_courses.xlsx', index=False)
```

```python
combined_data = pandas.concat([filtered_ai_data, filtered_com_data], ignore_index=True)

```

```python

# importing geopy library
from geopy.geocoders import Nominatim
import time

# calling the Nominatim tool
loc = Nominatim(user_agent="GetLoc")

# entering the location name

for location in combined_data["university_name_en"]:
    time.sleep(1)
    getLoc = loc.geocode(str(location))
    print(getLoc.address)

    combined_data["latitude"] = getLoc.latitude
    combined_data["longitude"] = getLoc.longitude 

# printing address

# printing latitude and longitude
# print("Latitude = ", getLoc.latitude, "\n")
# print("Longitude = ", getLoc.longitude)
```

```python
# Export combined data to Excel]
# combined_data.to_csv('data/filtered_ai_com_courses.csv', index=False)
combined_data.to_excel('data/filtered_ai_com_courses.xlsx', index=False)
```
